```python
import sempy
import sempy.fabric as fabric
```

```python
workspaceId = "e926f959-aad5-4096-beab-1f44d8bbec66"
datasetId = "f5f08b39-26fe-4832-be1b-0be430db5e40"
semantic_model_name = "AdventureWorksDW2020"
dax_query = "EVALUATE INFO.CALCDEPENDENCY()"

df = fabric.evaluate_dax(datasetId, dax_string = dax_query, workspace= workspaceId )
```
